## Módszertan

Az **EBITDA** (kamatok, adózás és értékcsökkenési leírás előtti eredmény) pozitív értéke azt jelzi, hogy az üzem mezőgazdasági alaptevékenysége jövedelmező, míg a negatív érték a tevékenység veszteséges működésére utal.  
Mindez az amortizáció, a pénzügyi és az adózási tételek figyelembevétele előtt értendő.

Ez a mutató különösen alkalmas **mezőgazdasági üzemek likviditási helyzetének értékelésére**, mivel a pénzügyi és beruházási tevékenységektől elválasztva mutatja meg a főtevékenységből származó pénzáramlást.  
Míg az adózás előtti és adózott eredmény számos torzító tényezőt figyelembe vesz, addig az EBITDA jobban tükrözi a főtevékenység gazdasági teljesítményét.

---

### Számítási képlet

$$
\text{EBITDA} = 1000 \times (\text{m4599\_03} + \text{m4559\_03})
$$

ahol:

- **m4599_03** – Üzemi tevékenység eredménye *(ezer Ft)*  
- **m4559_03** – Értékcsökkenési leírás összesen *(ezer Ft)*

---

### Adat- és árfolyamforrások

A számítások a **2023. és 2024. évi FADN-adatokon** alapulnak.  
A kifizetések átszámítása az **Európai Központi Bank** szeptember 30-án közzétett árfolyamai alapján történt:

- 2023-ban: **389,50 Ft/€**  
- 2024-ben: **396,88 Ft/€**

---

### DABIS-támogatás és szabályozási háttér

A **DABIS** esetében:

- **243,34 euró/ha** általános támogatás  
- **90,00 euró/ha** fiatal gazdálkodói támogatási rész  

Erre alkalmaztuk a **COM(2025) 560 bizottsági javaslat 6. cikk (3) bekezdésében** meghatározott *sávos támogatáscsökkentési (degresszív capping)* szabályokat.

---

### Eredmények értelmezése

A táblázatok tartalmazzák:

- a **súlyozatlan (nyers)** eredményeket, és  
- az **üzemi súllyal korrigált (megbízhatóbb)** eredményeket is.

---

### Mutatók értelmezése

- A **veszteséges üzemek arányának változása** azt mutatja, hány százalékponttal módosulna a jelenleg veszteséges üzemek aránya a fenti paraméterekkel bevezetett DABIS-hatás mellett.  
- Az **EBITDA-csökkenést mutató üzemek aránya** ezzel szemben az adott termelési irány és üzemméret-kategória egészének várható működési eredmény-változását jelzi előre.

```python
def young_farmer_area_filter_below(df: pd.DataFrame) -> pd.Series:
    mask_young = (df["young_farmer"] == 1) & interval_mask(
        df, Interval((0, 125.33), "right_closed")
    )
    mask_not_young = (df["young_farmer"] == 0) & interval_mask(
        df, Interval((0, 279.39), "right_closed")
    )
    return mask_young | mask_not_young


metrics = [
    "ebitda_ft",
]
pretty = {
    "ebitda_ft": "EBITDA (Ft)",
}
pivot_all = summarize_multi_metrics_pivot_with_interval(
    data, metrics, None, pretty_names=pretty, row_filter=young_farmer_area_filter_below
)
pivot_all.to_clipboard()
```

```python
def young_farmer_area_filter_over(df: pd.DataFrame) -> pd.Series:
    mask_young = (df["young_farmer"] == 1) & interval_mask(
        df, Interval((125.33, 50000.0), "right_closed")
    )
    mask_not_young = (df["young_farmer"] == 0) & interval_mask(
        df, Interval((279.39, 50000.0), "right_closed")
    )
    return mask_young | mask_not_young


metrics = [
    "ebitda_ft",
]
pretty = {
    "ebitda_ft": "EBITDA (Ft)",
}
pivot_all = summarize_multi_metrics_pivot_with_interval(
    data, metrics, None, pretty_names=pretty, row_filter=young_farmer_area_filter_over
)
pivot_all.to_clipboard()
```

```python
tbl_below_01 = prepare_base_df(
    data,
    metric="ebitda_ft",
    group_cols=("ev", "tip_m10ste"),
    interval=None,
    row_filter=young_farmer_area_filter_below,
)

tbl_below_01["tip_m10ste"] = pd.Categorical(
    tbl_below_01["tip_m10ste"], categories=ROW_ORDER[:-1]
)
tbl_below_01["ebitda_ft"] = tbl_below_01["weights_ste"] * tbl_below_01["ebitda_ft"]
tbl_below_01["ebitda_ft_dabis"] = (
    tbl_below_01["weights_ste"] * tbl_below_01["ebitda_ft_dabis"]
)

tbl_below_01 = tbl_below_01.groupby("tip_m10ste", observed=False).agg(
    ebitda_ft_sum=("ebitda_ft", "sum"), ebitda_ft_dabis_sum=("ebitda_ft_dabis", "sum")
)

tbl_below_01["ebitda_valtozas"] = 100 * (
    tbl_below_01["ebitda_ft_dabis_sum"] / tbl_below_01["ebitda_ft_sum"] - 1
)

tbl_below_02 = prepare_base_df(
    data,
    metric="ebitda_ft",
    group_cols=("ev",),
    interval=None,
    row_filter=young_farmer_area_filter_below,
)

tbl_below_02["ebitda_ft"] = tbl_below_02["weights_ste"] * tbl_below_02["ebitda_ft"]
tbl_below_02["ebitda_ft_dabis"] = (
    tbl_below_02["weights_ste"] * tbl_below_02["ebitda_ft_dabis"]
)

tbl_below_02 = tbl_below_02.groupby("ev").agg(
    ebitda_ft_sum=("ebitda_ft", "sum"), ebitda_ft_dabis_sum=("ebitda_ft_dabis", "sum")
)

tbl_below_02["ebitda_valtozas"] = 100 * (
    tbl_below_02["ebitda_ft_dabis_sum"] / tbl_below_02["ebitda_ft_sum"] - 1
)
tbl_below_02.index = ["teljes_mgi"]

tbl_below = pd.concat([tbl_below_01, tbl_below_02], axis=0)
tbl_below.loc[ROW_ORDER].to_clipboard()
```

```python
tbl_below_01 = prepare_base_df(
    data,
    metric="ebitda_ft",
    group_cols=("ev", "tip_m10ste"),
    interval=None,
    row_filter=young_farmer_area_filter_below,
)


tbl_below_01["tip_m10ste"] = pd.Categorical(
    tbl_below_01["tip_m10ste"], categories=ROW_ORDER[:-1]
)


eps = 1e-8
mask = tbl_below_01["ebitda_ft"].abs() > eps

tbl_below_01.loc[mask, "ebitda_valtozas_pct"] = 100 * (
    (tbl_below_01.loc[mask, "ebitda_ft_dabis"] - tbl_below_01.loc[mask, "ebitda_ft"])
    / tbl_below_01.loc[mask, "ebitda_ft"].abs()
)


def weighted_mean(group: pd.DataFrame) -> float:
    g = group.dropna(subset=["ebitda_valtozas_pct"])
    if g.empty:
        return np.nan
    return np.average(g["ebitda_valtozas_pct"], weights=g["weights_ste"])


tbl_below_01_agg = (
    tbl_below_01.groupby("tip_m10ste", observed=False)
    .apply(weighted_mean, include_groups=False)
    .rename("ebitda_valtozas_atlag_pct")
    .to_frame()
)


g_all = tbl_below_01.dropna(subset=["ebitda_valtozas_pct"])
belowall = np.average(g_all["ebitda_valtozas_pct"], weights=g_all["weights_ste"])

tbl_below_02 = pd.DataFrame(
    {"ebitda_valtozas_atlag_pct": [belowall]},
    index=["teljes_mgi"],
)

tbl_below = pd.concat([tbl_below_01_agg, tbl_below_02], axis=0)
tbl_below = tbl_below.loc[ROW_ORDER]
tbl_below.to_clipboard()
```

```python
# KDE for both series
kde1 = gaussian_kde(s1)
kde2 = gaussian_kde(s2)

x = np.linspace(0, 1000, 500)
y1 = kde1(x)
y2 = kde2(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label="s1 KDE", color="blue")
plt.plot(x, y2, label="s2 KDE", color="orange")

# Ratio
plt.figure(figsize=(8, 3))
plt.plot(x, y1 / y2, color="purple")
plt.axhline(1, color="grey", linestyle="--")
plt.xlabel("Mezőgazdasági terület (ha)")
plt.ylabel("KDE(s1) / KDE(s2)")
plt.title("Relatív reprezentativitás")
plt.grid(True)
plt.show()
```

```python
bins = np.linspace(0, 1000, 21)  # e.g. 50-ha bins
hist1, _ = np.histogram(s1, bins=bins, density=True)
hist2, _ = np.histogram(s2, bins=bins, density=True)

ratio = np.divide(hist1, hist2, out=np.full_like(hist1, np.nan), where=hist2 > 0)
bin_centers = (bins[:-1] + bins[1:]) / 2

plt.figure(figsize=(8, 4))
plt.plot(bin_centers, ratio, marker="o", color="purple")
plt.axhline(1, color="grey", linestyle="--")
plt.xlabel("Mezőgazdasági terület (ha)")
plt.ylabel("Mintaarány / Populációarány")
plt.title("Mintareprezentativitás (binned)")
plt.grid(True)
plt.show()
```
