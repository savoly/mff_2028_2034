# Degresszív támogatásra vonatkozó 130-240 EUR/ha értelmezése – három lehetséges megközelítés

## **1️⃣ Gazdaságszintű korlátozás**

- Példa: 1000 ha → maximális támogatás **100 000 EUR**  
- Ez **átlagosan 100 EUR/ha**, ami **alatta van a jogszabály által előírt 130–240 EUR/ha** sávnak.  
- **Probléma:** A nagyobb gazdaságok támogatása a kötelező átlagos minimum alá csökkenne, ami **ellentétes az uniós követelménnyel** és torzítaná a támogatási rendszer arányosságát.

---

## **2️⃣ Országos szint a flatrate támogatásra**

- **Szétírás nélkül 1 000-ha-ra:**  
  `apply_degressive_capping(240 * 1000) = 96 250 EUR`  
- **Szétírás 2×500 ha-ra:**  
  `2 × apply_degressive_capping(240 * 500) = 132 500 EUR`

- **Szétírás nélkül 10 000-ha-ra:**  
  `apply_degressive_capping(240 * 10_000) = 100 000 EUR`  
- **Szétírás 10 × 1000 ha-ra:**  
  `10 × apply_degressive_capping(240 * 1000) = 962 500 EUR`  
- **Probléma:** A farmok mesterséges szétírása **megnöveli az összkifizetést**, így a teljes országos támogatási boríték **kiszámíthatatlanná válik**.

**CAP regulation - Article 6. 2. – Young Farmers support remark**

Members States shall increase the support per eligible hectare granted to young farmers.*

**CAP regulation – Article 6.3 – Degressive Capping Rules**

(a) Member States shall reduce the annual amount of the area-based income support exceeding **EUR 20 000** to be granted to a farmer by **25%**  
  where the amount of the area-based income support granted to a farmer is between **EUR 20 000** and **EUR 50 000**.

(b) Member States shall reduce the annual amount of the area-based income support exceeding **EUR 50 000** to be granted to a farmer by **50%**  
  where the amount of the area-based income support granted to a farmer is more than **EUR 50 000** and not more than **EUR 75 000**.

(c) Member States shall reduce the annual amount of the area-based income support exceeding **EUR 75 000** to be granted to a farmer by **75%**  
  where the amount of the area-based income support granted to a farmer exceeds **EUR 75 000**.

---

**CAP regulation – Article 6.4 – Maximum Aid Limit**

The total amount of area-based income support shall not be higher than **EUR 100 000** per farmer per year.  
In the case of a legal person or groups of legal persons, the capping shall cover all holdings under the control of one legal or natural person.
